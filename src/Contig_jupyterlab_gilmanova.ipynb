```python
#contig.txt - файл со всеми найденными контигами
f = open('contig.txt', 'r')
f2 = open('contig_2.txt', 'w')

def chislo (line):
    return line.find('>seq')
kol_contig = 0
for line in f:
    if (chislo(line)!=-1):
        f2.write(line)
        kol_contig += 1
print('Количество контигов =', kol_contig)

f.close()
f2.close()

f2 = open('contig_2.txt', 'r')
length_contig = 0
max_length = 0
s = ''
list_length_0 = []
for line in f2:
    s = line
    l = ''
    for i in range (line.find('n')+1, line.find('c')-1):
        l += s[i]
    list_length_0.append(int(l))
    length_contig += int(l)
    if (int(l) > max_length):
        max_length = int(l)

print('Общая длина =', length_contig)
print('Длина самого большого контига =', max_length)

sum=0
N50=0
list_length = sorted(list_length_0, reverse=True)
for i in range (0, 613): 
    if (sum < length_contig*0.5):
        sum += list_length[i]
        N50 = list_length[i]
    else:
        break
print('N50 =',N50)
        
f2.close()
```
