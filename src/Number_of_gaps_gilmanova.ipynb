```python
f = open('longest.txt', 'r')
s = ''
num_gap = 0
sum_gap = 0
for line in f:
    s = line
    for i in range (len(s)):
        if (s[i]=='N'):
            sum_gap += 1
    for i in range (len(s)-1):
        if ((s[i]=='N') and (s[i+1]!='N')):
                num_gap+=1

print('Всего гэпов =', num_gap)
print('Общая длина гэпов =', sum_gap)
f.close()
```
